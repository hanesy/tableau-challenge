```python
#import dependencies
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
```

```python
# timeframe starts in January 2019
start_month = 1
start_year = 2019

# timeframe is for 12 months
months = 12

# array to hold the file name strings
months_string = []

# for loop to iterate through months to create appropriate file string format
for i in range(months):
    if i < 9:
        month = f"0{i+start_month}"
    else:
        month = f"{i+start_month}"
    months_string.append(month)
```

```python
# array to hold dataframes    
dataframes = []

# load csvs to dataframes and append the dataframes array
for month in months_string:
    print (f"loading {month}")
    csv = f"csvdata_JC/JC-2019{month}-citibike-tripdata.csv"
    load_csv = pd.read_csv(csv)
    load_csv["month"] = f"{month}/{start_year}"
    dataframes.append(load_csv)

# review to make sure the columns are the same
matchstatus = []

for i in range(len(dataframes)-1):
    if dataframes[i].columns.all() == dataframes[i+1].columns.all():
        matchstatus.append(True)
        print ("columns match")
    else:
        matchstatus.append(False)
        print("columns do not match")

print ("-----------")

for df in dataframes:
    print(df.columns)
    
# concat the dataframes if columns all match.
if False in matchstatus:
    print ("columns did not match")
else:
    print ("concat started")
    combined_df_JC = pd.concat(dataframes)
    combined_df_JC["location"] = "Jersey City"
    
print ("creating CSV")
combined_df_JC.to_csv("2019_citybike_JC.csv")
```
